```python
import pandas as pd

%matplotlib inline
import matplotlib
```

```python
imu1_data_file = '../sample_data/imu1_20181027_0954.txt'
imu2_data_file = '../sample_data/imu2_20181027_0954.txt'
```

```python
imu1 = pd.read_csv(imu1_data_file)
imu1.columns=imu1.columns.str.replace('#','')

imu2 = pd.read_csv(imu2_data_file)
imu2.columns=imu2.columns.str.replace('#','')

print(imu1.columns)
print(imu2.columns)
```

```python
zero_time = min( imu1['system_time_sec'].min(), imu2['system_time_sec'].min())

imu1['rel_time_sec'] = imu1['system_time_sec']-zero_time
imu2['rel_time_sec'] = imu2['system_time_sec']-zero_time

imu1.set_index('rel_time_sec');
imu2.set_index('rel_time_sec');
```

```python
pt1 = imu1.plot(x='rel_time_sec',y=['euler1', 'euler2', 'euler3'], figsize=(10,10))
pt1.set_xlabel( 'time (sec)' )
pt1.set_ylabel( 'radians' )
pt1.set_title('imu1')

pt2 = imu2.plot(x='rel_time_sec',y=['euler1', 'euler2', 'euler3'], figsize=(10,10))
pt2.set_xlabel( 'time (sec)' )
pt2.set_ylabel( 'radians' )
pt2.set_title('imu2')

```
